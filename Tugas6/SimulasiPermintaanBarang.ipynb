```python
import numpy as np
import matplotlib.pyplot as plt

# Mengatur parameter distribusi Poisson
lambda_ = 20  # Rata-rata permintaan barang per hari

# Jumlah simulasi (30 hari)
num_days = 30

# Melakukan simulasi permintaan barang menggunakan distribusi Poisson
daily_demand = np.random.poisson(lam=lambda_, size=num_days)

# Menampilkan hasil simulasi
print("Permintaan barang setiap hari:", daily_demand)

# Visualisasi hasil simulasi
plt.hist(daily_demand, bins=10, color='blue', edgecolor='black', alpha=0.7)
plt.title('Distribusi Permintaan Barang di Toko per Hari')
plt.xlabel('Jumlah Permintaan')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

```python
# Menghitung rata-rata hasil simulasi
simulated_mean = np.mean(daily_demand)
print(f"Rata-rata permintaan (simulasi): {simulated_mean}")
print(f"Nilai teoritis lambda: {lambda_}")
```
